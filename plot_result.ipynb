```python
from matplotlib import pyplot as plt
import json
```

```python
with open('losses.json') as f:
    data = json.load(f)

x = [i for i in range(len(data))]
y = [data[i] for i in range(len(data))]
plt.plot(x, y)
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```

```python
# Load the new JSON data
with open('his.json') as f:
    new_data = json.load(f)

# Extract success rate and epsilon
success_rate = [item['success'] / item['total'] for item in new_data]
epsilon = [item['espilon'] for item in new_data]

# Plot success rate
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(success_rate, label='Success Rate')
plt.xlabel('Index')
plt.ylabel('Success Rate')
plt.title('Success Rate Over Time')
plt.legend()

# Plot epsilon
plt.subplot(1, 2, 2)
plt.plot(epsilon, label='Epsilon', color='orange')
plt.xlabel('Index')
plt.ylabel('Epsilon')
plt.title('Epsilon Over Time')
plt.legend()

plt.tight_layout()
plt.show()
```
